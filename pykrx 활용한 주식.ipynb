# 4.1 종목명 붙이기

```python
import time
```

```python
stock_code = stock.get_market_ticker_list() # 현재일자 기준 가장 가까운 영업일의 코스피 상장종목 리스트
# stock_code
```

```python
final = pd.DataFrame()
for ticker in stock_code:
    df = stock.get_market_ohlcv_by_date(fromdate="20210830", todate="20210831", ticker=ticker)
    df = df.assign(종목코드=ticker, 종목명=stock.get_market_ticker_name(ticker))
    final = pd.concat([final, df], axis=0)
    time.sleep(0.2)
    print(f'{stock.get_market_ticker_name(ticker)} 완료')
final = final.reset_index()
final
```

```python
final.to_csv('save.csv')
```

# 9. 8월 30일 기준 전체 종목코드에 대해 모든 정보 합치기

```python
df_ohlcv = stock.get_market_ohlcv_by_ticker('20210831',market='KOSPI')
df_fundamental = stock.get_market_fundamental_by_ticker('20210831',market='KOSPI')
df_cap = stock.get_market_cap_by_ticker('20210831', market = 'KOSPI')
df_change_yes = stock.get_market_price_change_by_ticker(fromdate = '20210830',todate = '20210831', market='KOSPI')
df_net_purchase = stock.get_market_net_purchases_of_equities_by_ticker(fromdate = '20210831',todate = '20210831', market='KOSPI')
```

```python
df_list = [df_ohlcv, df_fundamental,df_cap,df_change_yes]
```

```python
final_df = df_net_purchase
```

```python
for df in df_list:
    col = list(set(df.columns)-set(final_df.columns))
    final_df = pd.merge(final_df, df[col],how='inner', left_on ='티커', right_on='티커')
```
